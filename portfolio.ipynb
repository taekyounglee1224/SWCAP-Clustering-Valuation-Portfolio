#### Cluster Portfolio와 Random Portfolio 구성

```python
from scipy.optimize import minimize

# Minimum Variance Portfolio using scipy.optimize
def maximize_return_portfolio_scipy(returns: pd.DataFrame):
    expected_returns = returns.mean().values  # 각 자산의 기대 수익률
    num_assets = len(expected_returns)

    # 목적 함수: 수익률의 음수 (최대화 ↔ 최소화)
    def negative_portfolio_return(weights):
        return -weights @ expected_returns

    constraints = {'type': 'eq', 'fun': lambda w: np.sum(w) - 1}
    bounds = tuple((0, 1) for _ in range(num_assets))
    init_guess = np.repeat(1 / num_assets, num_assets)

    result = minimize(negative_portfolio_return, init_guess, method='SLSQP', bounds=bounds, constraints=constraints)
    return result.x
    
# Cluster 5 portfolio weights
cluster_5_returns_in = returns_in_sample[cluster_5].dropna()
cluster_5_weights = maximize_return_portfolio_scipy(cluster_5_returns_in)

# Cluster 5 weighted return time series (in-sample & out-sample)
cluster_5_returns_out = returns_out_sample[cluster_5].dropna()
cluster_5_in_sample_pf = cluster_5_returns_in @ cluster_5_weights
cluster_5_out_sample_pf = cluster_5_returns_out @ cluster_5_weights

# 수정된 버전: 날짜 교집합 제거 & in/out-sample 분리 계산

random_portfolios_returns_in = []
random_portfolios_returns_out = []
random_portfolios_weights = []

for stock_list in random_portfolios:
    # 수익률 데이터 추출
    subset_returns_in = returns_in_sample[stock_list].dropna()
    subset_returns_out = returns_out_sample[stock_list].dropna()

    # 유효성 검사: 너무 적은 종목/날짜면 skip
    if subset_returns_in.empty or subset_returns_out.empty:
        continue
    if subset_returns_in.shape[1] < 10 or subset_returns_out.shape[1] < 10:
        continue

    # 포트폴리오 비중 계산 (in-sample 기준)
    weights = maximize_return_portfolio_scipy(subset_returns_in)

    # 수익률 계산
    in_ret = subset_returns_in @ weights
    out_ret = subset_returns_out @ weights

    random_portfolios_weights.append(weights)
    random_portfolios_returns_in.append(in_ret)
    random_portfolios_returns_out.append(out_ret)

# 최종 확인
len(random_portfolios_returns_in), len(random_portfolios_returns_out)
```

```python
# CVaR 계산 시 예외 처리 추가 (tail 데이터가 없을 경우 NaN 반환)
def sharpe_ratio(returns, risk_free_rate=0.0):
    excess_returns = returns - risk_free_rate
    return np.mean(excess_returns) / np.std(excess_returns)


def sortino_ratio(returns, risk_free_rate=0.0):
    downside_returns = returns[returns < risk_free_rate]
    return np.mean(returns - risk_free_rate) / np.std(downside_returns)


def cvar(returns, alpha=0.05):
    if len(returns) == 0:
        return np.nan
    var_threshold = np.quantile(returns, alpha)
    tail_losses = returns[returns <= var_threshold]
    return tail_losses.mean() if len(tail_losses) > 0 else np.nan


def max_drawdown(returns):
    cumulative = (1 + returns).cumprod()
    peak = cumulative.cummax()
    drawdown = (cumulative - peak) / peak
    return drawdown.min()


def volatility(returns):
    return np.std(returns)


# Cluster 5 포트폴리오 성과지표 계산
cluster_5_metrics_in = {
    'Sharpe': sharpe_ratio(cluster_5_in_sample_pf),
    'Sortino': sortino_ratio(cluster_5_in_sample_pf),
    'CVaR': cvar(cluster_5_in_sample_pf),
    'MDD': max_drawdown(cluster_5_in_sample_pf),
    'Volatility': volatility(cluster_5_in_sample_pf)
}

cluster_5_metrics_out = {
    'Sharpe': sharpe_ratio(cluster_5_out_sample_pf),
    'Sortino': sortino_ratio(cluster_5_out_sample_pf),
    'CVaR': cvar(cluster_5_out_sample_pf),
    'MDD': max_drawdown(cluster_5_out_sample_pf),
    'Volatility': volatility(cluster_5_out_sample_pf)
}

# 랜덤 포트폴리오 성과 지표 계산
random_metrics_in = []
random_metrics_out = []

for i in range(len(random_portfolios_returns_in)):
    r_in = random_portfolios_returns_in[i]
    r_out = random_portfolios_returns_out[i]

    if r_in.empty or r_out.empty:
        continue

    metrics_in = {
        'Sharpe': sharpe_ratio(r_in),
        'Sortino': sortino_ratio(r_in),
        'CVaR': cvar(r_in),
        'MDD': max_drawdown(r_in),
        'Volatility': volatility(r_in)
    }

    metrics_out = {
        'Sharpe': sharpe_ratio(r_out),
        'Sortino': sortino_ratio(r_out),
        'CVaR': cvar(r_out),
        'MDD': max_drawdown(r_out),
        'Volatility': volatility(r_out)
    }

    random_metrics_in.append(metrics_in)
    random_metrics_out.append(metrics_out)

# 데이터프레임으로 정리
df_metrics_in = pd.DataFrame(random_metrics_in)
df_metrics_out = pd.DataFrame(random_metrics_out)
```
