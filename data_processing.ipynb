### 필요한 라이브러리

```python
import pandas as pd
import numpy as np
```

### csv 파일 블러오기

```python
# 절대경로로 csv 파일 불러오기
file_path = 'financial_summary.csv'
df = pd.read_csv(file_path)
df
```

### 예측값이 있는 행 삭제

```python
# '기업명' 열에서 '예측'이 포함된 행을 제거한다.
# ~ 부정 연산자이다(True -> False, False -> True)
# na=False는 결측치를 False로 처리해서 오류를 방지한다.
df = df[~df['기업명'].str.contains('예측', na=False)]

df
```

### 분석할 열만 남기고 남은 열 삭제

```python
# 열 이름에 지정된 키워드가 포함된 열만 남기고 나머지 열을 전부 삭제한다.
keywords = ['기업명', 'PBR', 'PER', 'BPS', 'EPS', '자본총계', '당기순이익', 'ROE', 'ROA', '영업이익률', '순이익률', '부채비율', '이자발생부채', 'FCF', '영업활동현금흐름', '현금배당수익률', '매출액', '영업이익', '발행주식수', '자본금', '자산총계']
df = df[[col for col in df.columns if any(keyword in col for keyword in keywords)]]

# '지배', '비지배', '발표기준'라는 키워드가 포함된 열을 삭제한다.
columns_to_drop = [col for col in df.columns if any(keyword in col for keyword in ['지배', '비지배', '발표기준'])]
df = df.drop(columns=columns_to_drop)

df
```

### Nan값이 있는 열 확인

```python
# 열 별로 nan값의 수를 확인한다.
df.isnull().sum()

# df.columns
```

### Nan값을 가지고 있는 기업 9개 분석에서 배제

```python
# 지정된 열들이 전부 NaN인 행을 필터링해서 삭제할 행을 파악한다.
columns_to_check = ['매출액', '영업이익', '당기순이익', '자산총계', '자본총계', '자본금', '영업활동현금흐름', 'FCF', '이자발생부채', '영업이익률', '순이익률', '부채비율', 'EPS(원)', 'BPS(원)']
nan_rows = df[df[columns_to_check].isnull().all(axis=1)]

nan_rows
```

```python
# '기업명' 열에서 특정 키워드가 포함된 행을 제거한다.
keywords_to_remove = ['라온텍', '파두', '제이엔비', '워트', '오픈엣지테크놀로지', '엘케이켐', '아이언디바이스']
df = df[~df['기업명'].str.contains('|'.join(keywords_to_remove), na=False)]

df
```

### 추가적인 분석 기업 배제

```python
# PER값을 채우기 위해 수집한 종목들의 종가를 살펴보니, 162개 종목 중 46개 종목은 2020~2024년 사이에 상장이 되지 않아 해당 주식들은 drop하기로 결정했다.
# '기업명' 열에서 특정 키워드가 포함된 행을 제거
# 해당 기업들은 크롤링 과정에서 상장한지 얼마 되지 않아 재무제표가 없어 크롤링 자체가 안된 기업들도 포함되어 있다.
keywords_to_remove = ['티엘비', '아이언디바이스', '아이에스티이', '제이아이테크', '시지트로닉스', '지오엘리먼트', '아이엠티', '에이엘티', '티에프이', '퓨릿', '엠디바이스', '와이씨켐', '네패스아크', '지앤비에스 에코', '오픈엣지테크놀로지',
                      '자람테크놀로지', '티이엠씨', '기가비스', '큐알티', '위드텍', '넥스트칩', '코스텍시스템', '하이딥', '에이직랜드', '오로스테크놀로지', '그린리소스', '퀄리타스반도체', '샘씨엔에스', '마이크로투나노', '워트', '라온텍',
                      '넥스틴', '솔브레인', '저스템', 'HPSP', '가온칩스', '삼양엔씨켐', '엘케이켐', '비씨엔씨', '파두', '미래반도체', '레이저쎌', '사피엔반도체', '코스텍시스', '제이엔비', '티엘엔지니어링']
df = df[~df['기업명'].str.contains('|'.join(keywords_to_remove), na=False)]

df
```

```python
# 현재 '현금배당수익률'열의 Nan값이 상당히 많고, 해당 열의 값을 채우기 위한 정보가 부족하다.
# '현금배당수익률'열을 제거한다.
df = df.drop(columns=['현금배당수익률'])

df
```

### 'PER' 열 Nan값 채우기

```python
# 크롤링한 종가 데이터 csv 불러오기
# 절대경로 사용
df_price = pd.read_csv('stocks_year_end.csv')
```

```python
# 1. 'PER(배)'가 NaN인 행만 선택
df_nan_per = df[df['PER(배)'].isna()].copy()

# 2. 기업명과 연도 분리 (예: '아이앤씨(2020/12)' → 기업명: 아이앤씨, 연도: 2020)
df_nan_per[['기업명_추출', '연도']] = df_nan_per['기업명'].str.extract(r'(.+)\((\d{4})')

# 3. 연도에 맞는 종가 찾기: df_stock에서 해당 연도, 해당 기업명의 값을 가져와서 나눔
def get_stock_price(row):
    year = int(row['연도'])
    company = row['기업명_추출']
    try:
        # 연도별로 행이 2020년은 index 0부터 시작 → 2020: index 0, 2021: index 1, ...
        stock_price = df_price.loc[year - 2020, company]
        return stock_price
    except:
        return np.nan

# 주가 구하기
df_nan_per['주가'] = df_nan_per.apply(get_stock_price, axis=1)

# PER 계산: 주가 / EPS
df_nan_per['계산_PER'] = df_nan_per['주가'] / df_nan_per['EPS(원)']

# 원래 df_financial에 채워넣기
for idx, row in df_nan_per.iterrows():
    if not np.isnan(row['계산_PER']):
        df.at[idx, 'PER(배)'] = row['계산_PER']

# 수정된 일부 확인
df[df['기업명'].str.contains('아이앤씨')]  # 예시로 아이앤씨만 확인해봄
```

#### 'PBR' NaN 값 채우기

```python
# 1. 'PBR(배)'가 NaN인 행만 선택
df_nan_pbr = df[df['PBR(배)'].isna()].copy()

# 2. 기업명과 연도 분리 (예: '아이앤씨(2020/12)' → 기업명: 아이앤씨, 연도: 2020)
df_nan_pbr[['기업명_추출', '연도']] = df_nan_pbr['기업명'].str.extract(r'(.+)\((\d{4})')

# 3. 연도에 맞는 종가 찾기: df_price에서 해당 연도, 해당 기업명의 값을 가져와서 사용
def get_stock_price(row):
    year = int(row['연도'])
    company = row['기업명_추출']
    try:
        stock_price = df_price.loc[year - 2020, company]
        return stock_price
    except:
        return np.nan

# 주가 구하기
df_nan_pbr['주가'] = df_nan_pbr.apply(get_stock_price, axis=1)

# PBR 계산: 주가 / BPS
df_nan_pbr['계산_PBR'] = df_nan_pbr['주가'] / df_nan_pbr['BPS(원)']

# 원래 df에 채워넣기
for idx, row in df_nan_pbr.iterrows():
    if not np.isnan(row['계산_PBR']):
        df.at[idx, 'PBR(배)'] = row['계산_PBR']

# (선택) 수정된 일부 확인
df[df['기업명'].str.contains('아이앤씨')]  # 예시로 아이앤씨만
```

#### 'ROE, ROA' NaN 값 채우기

```python
# 1. 'ROE(%)'가 NaN인 행만 선택
df_nan_roe = df[df['ROE(%)'].isna()].copy()

# 2. 기업명과 연도 분리
df_nan_roe[['기업명_추출', '연도']] = df_nan_roe['기업명'].str.extract(r'(.+)\((\d{4})')

# 3. ROE 계산: (당기순이익 / 자본총계) * 100
df_nan_roe['계산_ROE'] = (df_nan_roe['당기순이익'] / df_nan_roe['자본총계']) * 100

# 원래 df에 채워넣기
for idx, row in df_nan_roe.iterrows():
    if not np.isnan(row['계산_ROE']):
        df.at[idx, 'ROE(%)'] = row['계산_ROE']

# 4. 'ROA(%)'가 NaN인 행만 선택
df_nan_roa = df[df['ROA(%)'].isna()].copy()

# 5. 기업명과 연도 분리
df_nan_roa[['기업명_추출', '연도']] = df_nan_roa['기업명'].str.extract(r'(.+)\((\d{4})')

# 6. ROA 계산: (당기순이익 / 자산총계) * 100
df_nan_roa['계산_ROA'] = (df_nan_roa['당기순이익'] / df_nan_roa['자산총계']) * 100

# 원래 df에 채워넣기
for idx, row in df_nan_roa.iterrows():
    if not np.isnan(row['계산_ROA']):
        df.at[idx, 'ROA(%)'] = row['계산_ROA']

# (선택) 수정된 일부 확인
df[df['기업명'].str.contains('아이앤씨')]
```
